```python
def load_bw_jsons(basepath, part):
    jsons = []
    with open(os.path.join(basepath, part + '.json'), 'rt') as f:
        for line in f:
            jsons.append(json.loads(line))
    return os.path.join(basepath, 'imgs'), jsons
```

```python
v2dev = load_bw_jsons(bw_v2, 'devset')
mndev = load_bw_jsons(bw_mnist, 'devset')
rnddev = load_bw_jsons(bw_random, 'devset')
```

```python
def show_sbs(path, imageA, imageB):
    f, (ax1, ax2) = plt.subplots(1, 2, sharey=True, figsize=(20,20))
    f.set
    ax1.imshow(plt.imread(os.path.join(path, imageA)))
    ax2.imshow(plt.imread(os.path.join(path, imageB)))
```

```python
show_sbs(bw_v2_imgs, 'bottcac._00.png', 'bottcac._01.png')
```

```python
def show_transition(bw, game, step):
    imgpath, injsons = bw
    injson = injsons[game]
    images = injson['images']
    this_trans = injson['notes'][step]
    imageA = images[this_trans['start']]
    imageB = images[this_trans['finish']]
    show_sbs(imgpath, imageA, imageB)
    print('\n'.join(this_trans['notes']))
```

```python
show_transition(v2dev, 0, 0)
```
